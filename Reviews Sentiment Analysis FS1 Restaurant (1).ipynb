```python
#import packages
import pandas as pd
import numpy as np
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
#reading the data
data = pd.read_excel('Reviews_FS1.xlsx')
print(data)
print(data.info())
```

```python
#creating the sentiment analysis object
sentiment = SentimentIntensityAnalyzer()
```

```python
#defining the function to analize the text
def analyze_sentiment(text):
    """
    Analyzes the sentiment of the given text.

    Parameters:
    - text (str): The input text for sentiment analysis.

    Returns:
    - str: Sentiment label ('positive', 'negative', 'neutral').
    """
    # Get the sentiment scores
    sentiment_scores = sentiment.polarity_scores(text)
 
    # Determine the sentiment label based on the compound score
    if sentiment_scores['compound'] >= 0.5:
        return 'positive'
    elif sentiment_scores['compound'] <= -0.5:
        return 'negative'
    else:
        return 'neutral'
```

```python
#analizing the comments in every row
list_sentiment = []

for idx, row in data.iterrows():
    list_sentiment.append(analyze_sentiment(row['Comment Text']))
    
```

```python
#show the list of the sentiment
list_sentiment
```

```python
#sentiment with indexes
leng = len(list_sentiment)
for i in range(leng):
    print(i,list_sentiment[i])
```

```python
#assigning the list in a new df column
data['Sentiments'] = list_sentiment
```
