```python
#read the data
cars_df = pd.read_csv('ecars.csv')
```

```python
cars_df
```

```python
cars_df_filtered = cars_df.loc[:9]
```

```python
#data with only countires
cars_df_filtered
```

```python
import re
def dropFootnoteAndCommas(text, pattern ='\[[^\]]*\]|,'):
  regex = re.compile(pattern)
  return regex.sub('', text)
```

```python
#test string for the function
text = 'United States[i][ii], Ecuador[IV], India[V]'
print(dropFootnoteAndCommas(text))
```

```python
cars_df_filtered['Country'] = cars_df_filtered['Country'].apply(dropFootnoteAndCommas)
```

```python
cars_df_filtered['Sales2021'] = cars_df_filtered['Sales2021'].apply(dropFootnoteAndCommas)
cars_df_filtered['Sales2020'] = cars_df_filtered['Sales2020'].apply(dropFootnoteAndCommas)
```

```python
cars_df_filtered
```

```python
cars_df_filtered['Sales2020'] = cars_df_filtered['Sales2020'].astype(int)
cars_df_filtered['Sales2021'] = cars_df_filtered['Sales2021'].astype(int)
```

```python
cars_df_filtered.info()
```

```python
#save the dataframe
cars_df_filtered.to_csv('ecars_cleaned.csv', index=False)
```
